## 1. The LASSO and Boosting for Regression

In this notebook, we will make an effort towards learning the concepts of Lasso and Boosting and implementing it on the socio-economic data from the 1990 US Census, law enforcement data from the 1990 US LEMAS survey, and crime data from the 1995 FBI UCR.

Import packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.tree import _tree
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, roc_curve, auc, accuracy_score,classification_report
from sklearn.linear_model import RidgeCV,Ridge,LassoCV,Lasso
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
import pprint
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

print("========== Packages Imported Successfully ==========\n")
```

### (a) Obtain Data

Download the Communities and Crime data from https://archive.ics.uci.edu/ml/datasets/Communities+and+Crime   
Use the first 1495 rows of data as
the training set and the rest as the test set.

```python
data_path = "../data/communities.data"
col_name_path = "../data/communities.names"

# This data has too many columns, so cannot assign the column names manually, hence writing a function to do the same.
def get_column_headers_from_file(filename):
    headers = []
    file = open(filename)
    for line in file:
        if "@attribute" in line:
            headers.append(line.split(" ")[1])
    return headers;

headers = get_column_headers_from_file(col_name_path)
df = pd.read_csv(data_path, header=None, na_values='?')
df.columns = headers
print("Data shape:", df.shape)
df.head()
```

```python
def get_train_test_data(df):
    train_df = df.iloc[:1495, :]
    test_df = df.iloc[1495: , :]
    return (train_df, test_df)

train_df, test_df = get_train_test_data(df)
print("Train data shape:", train_df.shape, "\nTest data shape:", test_df.shape)
```

### (b) Missing values

The data set has missing values. Use a data imputation technique to deal with
the missing values in the data set. The data description mentions some features
are nonpredictive. Ignore those features.

```python
# The first 5 features are non-predictive as given in the names document, hence removing these featues.
df = df.iloc[:, 5:]
train_df = train_df.iloc[:, 5:]
test_df = test_df.iloc[:, 5:]

print("All data shape:", df.shape, "\nTrain data shape:", train_df.shape, "\nTest data shape:", test_df.shape)
```

```python
# Now for imputation, using mean as the data imputation technique.

# Identifying columns that need to be imputed.
cols_with_nulls = df.columns[df.isna().any()].tolist()

def impute(df):
    imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
    for col in cols_with_nulls:
        imputer = imputer.fit(train_df[[col]])
        train_df[[col]] = imputer.transform(train_df[[col]])
        test_df[[col]] = imputer.transform(test_df[[col]])
    df_imputed = pd.concat([train_df,test_df])
    return df_imputed

df_imputed = impute(df)
print("Imputed data shape:", df.shape)
df_imputed.head()
```

### (c) Plot a correlation matrix

Plot a correlation matrix for the features in the data set.

```python
# Using the seaborn heatmap to plot the correlation matrix.
plt.figure(figsize=(50, 50))
corr = df_imputed.corr()
sns.heatmap(corr, cmap="YlGnBu", annot=True)
plt.show()
```

### (d) Calculate the Coefficient of Variation CV

Calculate the Coefficient of Variation CV for each feature, where CV = s/m, in which s is sample standard deviation and m is sample mean.

```python
def calculate_cv_features():
    describe = df_imputed.describe()
    std = describe.loc['std']
    mean = describe.loc['mean']
    describe.loc['CV'] = std / mean
    return describe.loc[['std','mean','CV']]

cv_features = calculate_cv_features()
cv_features
```

### (e) Scatter plots and box plots for highest CV features

Pick b = $\sqrt{128}$ features with highest CV, and make scatter plots and box plots for them. Can you draw conclusions about significance of those features, just by the scatter plots?

```python
def get_top_cv_features():    
    print("========== Top Features with Highest CV ==========")
    end = int(np.sqrt(128))
    top_cv_features = cv_features.loc['CV'].sort_values(ascending=False)
    top_cv_features = top_cv_features[:end]
    return top_cv_features

top_cv_features = get_top_cv_features()
top_cv_features
```

```python
def make_scatter_pairplot(df):
    sns.pairplot(df[top_cv_features.index])
    plt.show()

make_scatter_pairplot(df_imputed)
```

```python
def plot_box(df):
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(top_cv_features.index):
        plt.subplot(6, 2, i + 1)
        sns.boxplot(x=col, data=df)
    plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.7)

plot_box(df_imputed)
```

### (f) Fit a linear model

Fit a linear model using least squares to the training set and report the test error.

```python
def get_X_Y(df):
    x_data = df.iloc[:, :-1]
    y_data = df.iloc[:, -1:]
    return x_data, y_data

train_X, train_Y = get_X_Y(train_df)
test_X, test_Y = get_X_Y(test_df)

def get_test_error(trainX, trainY, testX, testY):
    train_sm = sm.add_constant(trainX)
    model = sm.OLS(trainY, train_sm).fit()
    predicted_dataY = model.predict(train_sm)
    print((model.summary()))
    mse_train = mean_squared_error(trainY, predicted_dataY)
    
    test_sm = sm.add_constant(testX)
    predicted_dataY_test = model.predict(test_sm)
    mse_test = mean_squared_error(testY, predicted_dataY_test)
    return round(mse_train, 4), round(mse_test, 4)

train_mse, test_mse = get_test_error(train_X, train_Y, test_X, test_Y)
```

```python
data = {
    "Training MSE": [train_mse],
    "Testing MSE": [test_mse]
}
mse_df = pd.DataFrame(data)
mse_df.T
```

### (g) Fit a ridge regression model

Fit a ridge regression model on the training set, with Î» chosen by cross-validation.
Report the test error obtained.

```python
alphas = []
def get_sample_alphas():
    for i in range(-10, 11, 1):
        alphas.append(10**i)
    return alphas

ridge_reg = RidgeCV(alphas=get_sample_alphas(), cv=KFold(n_splits=5, random_state=1, shuffle=True), scoring='neg_mean_squared_error')
ridge_reg.fit(train_X, train_Y)
print("========== Model Information ==========\n", ridge_reg)

pred_test_Y = ridge_reg.predict(test_X)
chosen_alpha = ridge_reg.alpha_
test_mse = mean_squared_error(test_Y,pred_test_Y)
test_r2 = r2_score(test_Y,pred_test_Y)

stats = {"Chosen Alpha": [chosen_alpha], "Test MSE": [test_mse], "Test R2": [test_r2]}

stats_df = pd.DataFrame(stats)
stats_df.T
```

### (h) Fit a LASSO model

Fit a LASSO model on the training set, with Î» chosen by cross-validation. Report
the test error obtained, along with a list of the variables selected by the model.
Repeat with standardized features. Report the test error for both cases and
compare them.

```python
def get_standardized_stats(train_x, train_y, test_x, test_y, is_standardized=False):
    scaler = StandardScaler()
    lr = LassoCV(alphas=get_sample_alphas(), cv=KFold(n_splits=5, random_state=50, shuffle=True), max_iter=10000)
    
    if is_standardized:
        train_x = scaler.fit_transform(train_x)
        test_x = scaler.fit_transform(test_x)
        
    lr.fit(train_x, train_y)
    pred_test_y = lr.predict(test_x)

    chosen_alpha = lr.alpha_
    test_mse = mean_squared_error(test_y, pred_test_y)
    test_r2 = r2_score(test_y,pred_test_y)

    stats = {"Chosen Alpha": [chosen_alpha], "Test MSE": [test_mse], "Test R2": [test_r2]}

    stats_df = pd.DataFrame(stats)
    return stats_df.T, lr

lasso_normalized_df, lr = get_standardized_stats(train_X, train_Y, test_X, test_Y)
lasso_normalized_df
```

```python
def get_selected_variable_list():    
    lasso_dict_coef = {}
    count_cols = len(train_X.columns)
    for i in range(count_cols):    
        if lr.coef_[i] != 0:
            lasso_dict_coef[train_X.columns[i]] = [round(lr.coef_[i], 4)]

    print('Number of features: ', len(lasso_dict_coef))
    print('==============================================')
    pprint.pprint(lasso_dict_coef)

get_selected_variable_list()
```

### (i) Fit a PCR model

Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
def pcr_model(train_x, train_y, dimension):
    scores=[]
    CV = 5
    job_count = -1
    pca = PCA(n_components = dimension)
    LR = LinearRegression(fit_intercept=True)    
    pipeline = make_pipeline(pca, LR)
    scores=cross_val_score(pipeline, train_x, train_y, cv=CV, n_jobs=job_count, scoring="neg_mean_squared_error")
    
    return sum(abs(scores))/len(scores)

pcr_dict = {}
count_cols = len(train_X.columns)
for i in range(1,count_cols + 1):
    pcr_dict[i] = pcr_model(train_X, train_Y, i)

pprint.pprint(pcr_dict)
minval=float('inf')
for key,value in pcr_dict.items():
    if value < minval:
        min_key = key
    minval = min(minval, value)

pcr_stats = {"Selected M": [min_key], "Minimum Validation Error": [minval]}

pcr_df = pd.DataFrame(pcr_stats)
pcr_df
```

```python
def get_pca_test_error():    
    count_components = pcr_stats['Selected M'][0]
    pca = PCA(n_components = count_components).fit(train_X)
    pca_train = pca.transform(train_X)
    pca_test = pca.transform(test_X)
    LR = LinearRegression().fit(pca_train, train_Y)
    predicted_Y_test = LR.predict(pca_test)
    pca_test_mse = mean_squared_error(test_Y, predicted_Y_test)
    return round(pca_test_mse, 4)

print("PCA Test MSE: ", get_pca_test_error())
```

### (j) Fit a boosting tree

In this section, we would like to fit a boosting tree to the data. As in classification
trees, one can use any type of regression at each node to build a multivariate
regression tree. Because the number of variables is large in this problem, one
can use L1-penalized regression at each node. Such a tree is called L1 penalized
gradient boosting tree. You can use XGBoost5
to fit the model tree. Determine
Î± (the regularization term) using cross-vali.dation

```python
stats_xgb = {}
for alpha in get_sample_alphas():
    model = xgb.XGBRegressor(objective='reg:squarederror', reg_alpha = alpha, random_state = 50)
    score = cross_val_score(model, train_X, train_Y, cv=KFold(5, shuffle=True,random_state=50), n_jobs=-1, scoring="neg_mean_squared_error")
    
    val = sum(abs(score))/len(score)
    stats_xgb[alpha] = f'{round(val, 4)}'
    
pprint.pprint(stats_xgb)
minval=float('inf')

for key, value in stats_xgb.items():
    if float(value) < minval:
        minval = float(value)
        min_key = key

xgb_stats = {"Selected Alpha": [min_key], "Minimum Validation Error MSE": [minval]}
xgb_df = pd.DataFrame(xgb_stats)
xgb_df
```

```python
def get_xgb_test_error():
    model = xgb.XGBRegressor(objective='reg:squarederror', reg_alpha = xgb_stats['Selected Alpha'][0], random_state = 50)
    model.fit(train_X,train_Y)
    
    test_predicted_xgb = model.predict(test_X)
    test_mse = mean_squared_error(test_Y,test_predicted_xgb)
    
    return test_mse

print("XGBoost Test MSE: ", round(get_xgb_test_error(), 4))
```
